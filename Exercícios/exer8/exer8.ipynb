```python
import pandas as pd
import numpy as np
import waipy
import warnings
warnings.filterwarnings('ignore')
```

```python
nomeArquivo = 'Germany.txt'
nome = nomeArquivo.split('.')[0]+' COVID-19'
data = np.genfromtxt(nomeArquivo, dtype = 'float32',filling_values = 0)
# Removendo entradas iguais a zero
data = data[data != 0]

x = np.linspace(0, len(data), len(data))

# Data normalization
data_norm = waipy.normalize(data)

# Continuous wavelet transform
morlet = waipy.cwt(data_norm, dt=0.25, pad=1, dj=0.25, s0=2*0.25, j1=7/0.25, lag1=0.72, param=6, mother='Morlet', name=nome)

# Plotting the result
waipy.wavelet_plot(var=nome, time=x, data=data_norm, dtmin=0.03125, result=morlet)
```

```python
dog = waipy.cwt(data_norm, dt=0.25, pad=1, dj=0.25, s0=2*0.25, j1=7/0.25, lag1=0.72, param=2, mother='DOG', name=nome)
waipy.wavelet_plot(var=nome, time=x, data=data_norm, dtmin=0.03125, result=dog)
```
