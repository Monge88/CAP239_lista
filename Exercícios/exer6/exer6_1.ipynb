Este código foi desenvolvido em torno do algoritmo <font color='blue'>specplus.py</font> para o cálculo do valor de $\alpha$ e $\beta$ respectivos a cada um dos sinais do dataset_signal. Estes sinais são gerados automaticamente pelo próprio algoritmo com a importação dos módulos powernoise, pmodel, logistic e henon. De maneira geral o código realiza os seguintes processos:<br>
**1 -** Gera uma quantidade n de sinais para cada uma das séries. A quantidade de sinais e o número de pontos de cada um podem ser livremente alterados; <br>
**2 -** Calcula, via PSD e DFA, os valores de $\alpha$ e $\beta$ para cada um dos n sinais;<br>
**3 -** Grafica os valores de $\alpha$ e $\beta$ apenas para o primeiro sinal de cada série;<br>
**4 -** Faz e grafica a comparação entre os valores de $\beta$ a partir do PSD e da fórmula 2 * $\alpha$ - 1; <br>
**5 -** Cria os espaços e classifica os sinais gerados nos espaços de parâmetros EPSB-K-means: S² x K x $\beta$ e EDF-K-means: S² x K x $\alpha$ com o uso do algoritmo do k-means;<br>
**6 -** Identifica e mostra os sinais em cada um dos clusters criados pelo k-means.<br>

Um exemplo de execução é mostrado em seguida e alguns resultados discutidos ao final.

```python
#SPECPLUS.py
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
Created on Mon Nov 11 04:26:21 2019
@author: reinaldo
"""
__authors__ = 'Paulo Giovani & Reinaldo Rosa'
__copyright__ = 'Copyright 2017, 3DBMO Project INPE'
__credits__ = ['Paulo Giovani', 'Reinaldo Roberto Rosa', 'Murilo da Silva Dantas']
__license__ = 'GPL'
__version__ = '0.1B'
__maintainer__ = 'Paulo Giovani'
__email__ = 'pg_faria@yahoo.com.br'

# ------------------------------------------------------------------------
# Specplus.pyplot
# ------------------------------------------------------------------------

from powernoise import powernoise
from p_model import pmodel
from logistic import logistic
from henon import henon
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from mpl_toolkits.mplot3d import Axes3D
from scipy import stats, optimize
import numpy as np
import pandas as pd
import seaborn as sns
import random
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
import math
import csv
import os
import warnings
warnings.filterwarnings('ignore')

#---------------------------------------------------------------------

# Calcula o PSD da série temporal

#---------------------------------------------------------------------

def psd(data):

    """Calcula o PSD de uma série temporal."""
    # Define um intervalo para realizar o ajuste da reta
    INICIO = 10
    FIM = 200

    # O vetor com o tempo é o tamanho do número de pontos
    N = len(data)
    tempo = np.arange(len(data))

    # Define a frequência de amostragem
    dt = (tempo[-1] - tempo[0] / (N - 1))
    fs = 1 / dt

    # Calcula o PSD utilizando o MLAB
    power, freqs = mlab.psd(data, Fs = fs, NFFT = N, scale_by_freq = False)

    # Calcula a porcentagem de pontos utilizados na reta de ajuste
    totalFrequencias = len(freqs)
    totalPSD = FIM - INICIO
    porcentagemPSD = int(100 * totalPSD / totalFrequencias)
    
    # Seleciona os dados dentro do intervalo de seleção
    xdata = freqs[INICIO:FIM]
    ydata = power[INICIO:FIM]

    # Simula o erro
    yerr = 0.2 * ydata

    # Define uma função para calcular a Lei de Potência
    powerlaw = lambda x, amp, index: amp * (x**index)

    # Converte os dados para o formato LOG
    logx = np.log10(xdata)
    logy = np.log10(ydata)

    # Define a função para realizar o ajuste
    fitfunc = lambda p, x: p[0] + p[1] * x
    errfunc = lambda p, x, y, err: (y - fitfunc(p, x)) / err    
    logyerr = yerr / ydata

    # Calcula a reta de ajuste
    pinit = [1.0, -1.0]
    out = optimize.leastsq(errfunc, pinit, args = (logx, logy, logyerr), full_output = 1)    
    pfinal = out[0]
    covar = out[1]
    index = pfinal[1]
    amp = 10.0 ** pfinal[0]
    indexErr = np.sqrt(covar[0][0])
    ampErr = np.sqrt(covar[1][1]) * amp

    # Retorna os valores obtidos
    return freqs, power, xdata, ydata, amp, index, powerlaw, INICIO, FIM

#---------------------------------------------------------------------

# Calcula o DFA 1D da série temporal

#---------------------------------------------------------------------

def dfa1d(timeSeries, grau):
    """Calcula o DFA 1D (adaptado de Physionet), onde a escala cresce
    de acordo com a variável 'Boxratio'. Retorna o array 'vetoutput', 
    onde a primeira coluna é o log da escala S e a segunda coluna é o
    log da função de flutuação."""

    # 1. A série temporal {Xk} com k = 1, ..., N é integrada na chamada função perfil Y(k)
    x = np.mean(timeSeries)
    timeSeries = timeSeries - x
    yk = np.cumsum(timeSeries)
    tam = len(timeSeries)

    # 2. A série (ou perfil) Y(k) é dividida em N intervalos não sobrepostos de tamanho S
    sf = np.ceil(tam / 4).astype(np.int)
    boxratio = np.power(2.0, 1.0 / 8.0)
    vetoutput = np.zeros(shape = (1,2))

    s = 4
    while s <= sf:        
        serie = yk        
        if np.mod(tam, s) != 0:
            l = s * int(np.trunc(tam/s))
            serie = yk[0:l]
            
        t = np.arange(s, len(serie), s)
        v = np.array(np.array_split(serie, t))
        l = len(v)
        x = np.arange(1, s + 1)

        # 3. Calcula-se a variância para cada segmento v = 1,…, n_s:
        p = np.polynomial.polynomial.polyfit(x, v.T, grau)
        yfit = np.polynomial.polynomial.polyval(x, p)
        vetvar = np.var(v - yfit)

        # 4. Calcula-se a função de flutuação DFA como a média das variâncias de cada intervalo
        fs = np.sqrt(np.mean(vetvar))
        vetoutput = np.vstack((vetoutput,[s, fs]))

        # A escala S cresce numa série geométrica
        s = np.ceil(s * boxratio).astype(np.int)

    # Array com o log da escala S e o log da função de flutuação   
    vetoutput = np.log10(vetoutput[1::1,:])

    # Separa as colunas do vetor 'vetoutput'
    x = vetoutput[:,0]
    y = vetoutput[:,1]

    # Regressão linear
    slope, intercept, _, _, _ = stats.linregress(x, y)

    # Calcula a reta de inclinação
    predict_y = intercept + slope * x

    # Calcula o erro
    pred_error = y - predict_y

    # Retorna o valor do ALFA, o vetor 'vetoutput', os vetores X e Y,
    # o vetor com os valores da reta de inclinação e o vetor de erros
    return slope, vetoutput, x, y, predict_y, pred_error


def k_means(df):
    
    # Variável auxiliar para separar os espaços de parâmetros
    tmp = ['alfa','beta']
    
    for n in tmp:
        
        # Definindo o nome de n atual
        if n == 'alfa':
            name = 'beta'
        elif n == 'beta':
            name = 'alfa'
            
        # Fazendo uma cópia do arquivo de entrada para renovar os valores a cada loop
        data = df.copy()
        
        # Retirando uma das colunas (alfa ou beta) para separar os espaços de parâmetros
        data = data.drop(n, axis=1)
              
        # Descobrindo o k ótimo
        model = KMeans()
        visualizer = KElbowVisualizer(model, k=10)
        visualizer.fit(data)
        plt.close()

        # Aplicando o k-means
        n_clusters = visualizer.elbow_value_ #Número ótimo de clusters
        kmeans = KMeans(n_clusters=n_clusters)
        kmeans.fit(data)
        data['K_classes'] = kmeans.labels_
        centroides = kmeans.cluster_centers_

        #Plotando as relações a pares de cada medida estatística (seaborn)
        colors = sns.color_palette('tab10',n_clusters)
        
        try:
            g = sns.pairplot(data, 'K_classes', palette=colors)
            g.fig.suptitle('\n\nEspaço de parâmetros: S² x K x '+name, y=1.08)
            plt.show()
        except ValueError:
            print('\nOs dados não apresentam variância')
        except RuntimeError:
            print('\nOs dados não apresentam variância')


        #Plotando os pontos em 3 dimensões
        fig = plt.figure(figsize=(20,15))
        ax = fig.add_subplot(111, projection='3d')

        labels = data.groupby('K_classes')
        for key, item in labels: 
            x = item['skewness²']
            y = item['kurtosis']
            z = item[name]


            ax.scatter(x, y, z, label=key, alpha=1, s=200, edgecolors='b', c=[colors[key]])


        #Plotandos os centróides em 3 dimensões
        x = []
        y = []
        z = []
        for row in centroides:
            x.append(row[0])
            y.append(row[1])
            z.append(row[2])

        ax.scatter(x,y,z, alpha=0.4, s=500, label='Centróides')

        #Ajustando os eixos
        ax.set_xlabel('\n\n'+data.columns[0],fontsize=20)
        ax.set_ylabel('\n\n'+data.columns[1],fontsize=20)
        ax.set_zlabel('\n\n'+data.columns[2],fontsize=20)
        ax.xaxis.set_tick_params(labelsize=18)
        ax.yaxis.set_tick_params(labelsize=18)
        ax.zaxis.set_tick_params(labelsize=18)

        plt.tight_layout()
        plt.legend(loc='best', fontsize=20) 
        plt.show()

        for classe in pd.unique(data['K_classes']):
            print(f'\n\n****************** Sinais na classe {classe} ******************')
            print(data.loc[data['K_classes']==classe])
    

def beta_check(beta_psd,beta_wkp,n,alfa):
    """
    Checa se os valores de beta estão bem ajustados à formula 2 * alfa - 1
    """
    
    # Separa o vetor com os valores de beta em chunks correspondentes a cada série
    chunks_psd = [beta_psd[x:x+n] for x in range(0,len(beta_psd),n)]
    chunks_wkp = [beta_wkp[x:x+n] for x in range(0,len(beta_wkp),n)]
    chunks_alfa = [alfa[x:x+n] for x in range(0,len(alfa),n)]
    sinais = ['Stocastic','White noise','Pink noise','Red noise','Endogenous','Exogenous','Logistic','Henon']
    
    plot_counter = 1
    fig = plt.figure(figsize=(14,14))
    
    for i in range(len(chunks_psd)):
        plt.subplot(int(np.ceil(len(sinais)/2)),2,plot_counter)
        x_wkp = chunks_alfa[i]
        y_wkp = chunks_wkp[i]
        x_psd = chunks_alfa[i]
        y_psd = chunks_psd[i]
        
        slope1, intercept1, r_value1, p_value1, std_err1 = stats.linregress(x_wkp,y_wkp)
        slope2, intercept2, r_value2, p_value2, std_err2 = stats.linregress(x_psd,y_psd)
        
        
        plt.plot(x_wkp, [slope1 * x + intercept1 for x in x_wkp], c='b', 
                 label=f'Beta (WKP): {slope1:.1f}\u03B1 + ({intercept1:.1f})')
        plt.plot(x_psd, [slope2 * x + intercept2 for x in x_psd], c='orange', 
                 label=f'Beta (PSD):  {slope2:.1f}\u03B1 + ({intercept2:.1f})')
        
        plt.scatter(x_wkp,y_wkp)
        plt.scatter(x_psd,y_psd)
        plt.title(sinais[i])
        plt.grid()
        plt.legend(loc='best')
        plot_counter = plot_counter + 1
        
    fig.text(0.5, 1.0, '\n\nComparação entre os valores de Beta (PSD) e Beta (WKP)', ha='center', 
             fontsize=16)  
    fig.text(0.5, 0.0, 'Alfa', ha='center', fontsize=14)
    fig.text(0.0, 0.5, 'Beta', va='center', rotation='vertical', fontsize=14)
    plt.tight_layout(pad=3.0, h_pad=1.0)
    plt.show()
    
    
        
def colname(n,name):
    """
    Retorna uma lista com as strings dos valores de N e sinal atuais
    """

    col = []
    for i in range(n):
        col.append(name+'_'+str(i))

    return col  
        
        
        
#---------------------------------------------------------------------

# Trecho principal

#---------------------------------------------------------------------

def main():

    """Função com o código principal do programa."""

    print("\nSpectral Analysis for time series...\n")

    # Desabilita as mensagens de erro do Numpy (warnings)
    old_settings = np.seterr(divide = 'ignore', invalid = 'ignore', over = 'ignore')

    n_iter = 10
    n_sinais = 2**13
    assimetria_ = []
    curtose_ = []
    beta_ = []
    alfa_ = []
    
    
    # Cria 10 sinais referentes às séries do dataset_signal
    S1 = []
    S2 = []
    S3 = []
    S4 = []
    S5 = []
    S6 = []
    S7 = []
    S8 = []

    for n in range(n_iter):
        #série temporal estocástica
        res = n_sinais/12
        S1.append((np.random.randn(n_sinais) * np.sqrt(res) * np.sqrt(1 / n_sinais)).cumsum())

        #ruído branco
        S2.append(powernoise(0,n_sinais))

        #ruído rosa
        S3.append(powernoise(1,n_sinais))

        #ruído vermelho
        S4.append(powernoise(2,n_sinais))

        #sinal endógeno
        S5.append(pmodel(n_sinais,0.4,-0.4))

        #sinal exógeno
        S6.append(pmodel(n_sinais,0.2,-0.7))

        #mapa logístico
        rho = np.random.uniform(3.81,4.00)
        S7.append(logistic(rho,0.001,n_sinais))

        #mapa henon
        a = np.random.uniform(1.35,1.42)
        S8.append(henon(a,0.3,0.1,n_sinais))

    tmp1 = pd.DataFrame(np.transpose(S1),columns=colname(n_iter,'Stocasthic noise'))
    tmp2 = pd.DataFrame(np.transpose(S2),columns=colname(n_iter,'White noise'))
    tmp3 = pd.DataFrame(np.transpose(S3),columns=colname(n_iter,'Pink noise'))
    tmp4 = pd.DataFrame(np.transpose(S4),columns=colname(n_iter,'Red noise'))
    tmp5 = pd.DataFrame(np.transpose(S5),columns=colname(n_iter,'Endogenous'))
    tmp6 = pd.DataFrame(np.transpose(S6),columns=colname(n_iter,'Exogenous'))
    tmp7 = pd.DataFrame(np.transpose(S7),columns=colname(n_iter,'Logistic'))
    tmp8 = pd.DataFrame(np.transpose(S8),columns=colname(n_iter,'Henon'))

    df = pd.concat([tmp1,tmp2,tmp3,tmp4,tmp5,tmp6,tmp7,tmp8],axis=1)

    # Dataframe com os momentos estatísticos dos sinas acima
    df_stats = pd.DataFrame(np.transpose([stats.skew(df)**2,stats.kurtosis(df)]),
                            columns=['skewness²','kurtosis']).set_index(df.columns)


    # Calcula o PSD e DFA para cada coluna
    for column in df.columns:
        
        data = df[column].values
           
       
        # Para cada série calcule S² e K
        assimetria_.append(stats.skew(data)**2)
        curtose_.append(stats.kurtosis(data))
        
        #-----------------------------------------------------------------

            # Cálculo do PSD

        #-----------------------------------------------------------------
        
        # Calcula o PSD
        freqs, power, xdata, ydata, amp, index, powerlaw, INICIO, FIM = psd(data)

        # O valor do beta equivale ao index 
        b = index
        beta_.append(b)
        
        #-----------------------------------------------------------------

            # Cálculo do DFA

        #-----------------------------------------colornoise.csv------------------------
        
        # Calcula o DFA 1D
        alfa, vetoutput, x, y, reta, erro = dfa1d(data, 1)
        alfa_.append(alfa)  
        
       
        # Plota apenas a primeira iteração de cada série
        if column.split('_')[1] == '0':
            # Exibe os primeiro N valores do arquivo
            N = 10
            print("\n\nOriginal time series data (%d points): \n" %(len(data)))
            print("First %d points: %s\n" %(N, data[0:10]))
            print()

            #-----------------------------------------------------------------

                # Parâmetros gerais de plotagem

            #-----------------------------------------------------------------

            # Define os subplots
            fig = plt.figure()
            fig.subplots_adjust(hspace = .3, wspace = .2)

            # Tamanho das fontes
            tamanhoFonteEixoX = 16
            tamanhoFonteEixoY = 16
            tamanhoFonteTitulo = 16
            tamanhoFontePrincipal = 25

            # Título principal

            tituloPrincipal = 'Time Series Spectral Analysis - Serie:'+column 

            #-----------------------------------------------------------------

                # Plotagem da série original

            #-----------------------------------------------------------------
            plt.style.use('default')
            # Define as cores da plotagem
            corSerieOriginal = 'r'

            # Título dos eixos da série original
            textoEixoX = 'Tempo'
            textoEixoY = 'Amplitude'
            textoTituloOriginal = 'Original Time Series Data'
            print("1. Plotting time series data...")

            # Plotagem da série de dados    
            #O = fig.add_subplot(1, 3, 1)    
            O = fig.add_subplot(2, 1, 1)
            O.plot(data, '-', color = corSerieOriginal)
            O.set_title(textoTituloOriginal, fontsize = tamanhoFonteTitulo)
            O.set_xlabel(textoEixoX, fontsize = tamanhoFonteEixoX)
            O.set_ylabel(textoEixoY, fontsize = tamanhoFonteEixoY)
            O.ticklabel_format(style = 'sci', axis = 'x', scilimits = (0,0))
            O.grid()

            #-----------------------------------------------------------------

                # Plotagem do PSD

            #-----------------------------------------------------------------


            # Define as cores da plotagem
            corPSD1 = 'k'
            corPSD2 = 'navy'

            # Título dos eixos do PSD
            textoPSDX = 'Frequência (Hz)'
            textoPSDY = 'Potência'
            textoTituloPSD = r'Power Spectrum Density $\beta$ = '
            print("2. Plotting Power Spectrum Density...")

            # Plotagem do PSD    
            PSD = fig.add_subplot(2, 2, 3)    
            PSD.plot(freqs, power, '-', color = corPSD1, alpha = 0.7)
            PSD.plot(xdata, ydata, color = corPSD2, alpha = 0.8)
            PSD.axvline(freqs[INICIO], color = corPSD2, linestyle = '--')
            PSD.axvline(freqs[FIM], color = corPSD2, linestyle = '--')    
            PSD.plot(xdata, powerlaw(xdata, amp, index), 'r-', linewidth = 1.5, label = '$%.4f$' %(b))    
            PSD.set_xlabel(textoPSDX, fontsize = tamanhoFonteEixoX)
            PSD.set_ylabel(textoPSDY, fontsize = tamanhoFonteEixoY)
            PSD.set_title(textoTituloPSD + '%.4f' %(b), loc = 'center', fontsize = tamanhoFonteTitulo)
            PSD.set_yscale('log')
            PSD.set_xscale('log')
            PSD.grid() 

            #-----------------------------------------------------------------

                # Plotagem do DFA

            #-----------------------------------------------------------------

            # Verifica se o DFA possui um valor válido
            # Em caso afirmativo, faz a plotagem
            if not math.isnan(alfa):

                # Define as cores da plotagem
                corDFA = 'darkmagenta'

                # Título dos eixos do DFA
                textoDFAX = '$log_{10}$ (s)'
                textoDFAY = '$log_{10}$ F(s)'
                textoTituloDFA = r'Detrended Fluctuation Analysis $\alpha$ = '
                print("3. Plotting Detrended Fluctuation Analysis...")

                # Plotagem do DFA 
                DFA = fig.add_subplot(2, 2, 4)    
                DFA.plot(x, y, 's', 
                         color = corDFA, 
                         markersize = 4,
                         markeredgecolor = 'r',
                         markerfacecolor = 'None',
                         alpha = 0.8)


                DFA.plot(x, reta, '-', color = corDFA, linewidth = 1.5)
                DFA.set_title(textoTituloDFA + '%.4f' %(alfa), loc = 'center', fontsize = tamanhoFonteTitulo)
                DFA.set_xlabel(textoDFAX, fontsize = tamanhoFonteEixoX)
                DFA.set_ylabel(textoDFAY, fontsize = tamanhoFonteEixoY)
                DFA.grid()


            else:  
                DFA = fig.add_subplot(2, 2, 4)
                DFA.set_title(textoTituloDFA + 'N.A.', loc = 'center', fontsize = tamanhoFonteTitulo)
                DFA.grid()


            #-----------------------------------------------------------------

            # Exibe e salva a figura

            #-----------------------------------------------------------------

            plt.suptitle(tituloPrincipal, fontsize = tamanhoFontePrincipal)
         #   nomeImagem = 'ANALYSIS_PSD_DFA_2.png'
            fig.set_size_inches(15, 9)
          #  plt.savefig(nomeImagem, dpi = 300, bbox_inches = 'tight', pad_inches = 0.1)
            plt.show() 
    
    # Calculando os valores de beta a partir de alfa
    beta_wkp = [2*n - 1 for n in alfa_]
    
    #Compara os valores de beta
    beta_check(beta_,beta_wkp,n_iter,alfa_)
    
    # Adicionando os valores de alfa e beta no dataframe do k-means
    df_stats['beta'] = beta_
    df_stats['alfa'] = alfa_
    
    # Retira os sinais endógeno e exógeno do ajuste com o k-means porque seus momentos são muito superiores
    # aos demais, o que impossibilita uma melhor análise do agrupammento dos sinais
    name_to_drop = []
    for signal in df_stats.index:
        if signal.split('_')[0] == 'Endogenous' or signal.split('_')[0] == 'Exogenous':
            name_to_drop.append(signal)
            
    df_stats = df_stats.drop(name_to_drop,axis=0)
        
    # Chamando o k-means
    k_means(df_stats)

    

#---------------------------------------------------------------------

# Trecho principal

#---------------------------------------------------------------------

if __name__ == "__main__":
    main()
```

Vale notar que para a maior parte do sinais, os valores de $\beta$ estão bem ajustados pela equação 2$\alpha$ - 1, lembrando que os valores de beta gerados pelo PSD estão com sinal negativo e por isso essa aparente simetria de valores na imagem abaixo.

Outro resultado interessante ocorreu no agrupamento do k-means. Os sinais foram alocados em grupos diferentes dependendo do espaço de parâmetros utilizado. Para o espaço de parâmetros S² x K x $\beta$, o algoritmo do k-means alocou as séries em cinco grupos distintos, exatamente o número de séries utilizada. No entanto este agrupamento não alocou cada série em grupos distintos. 

De fato, apenas o ruído branco e o ruído rosa mantiveram todos os seus sinais em um mesmo cluster. Os demais tiveram seus sinais alocados em grupos diferentes sem nenhum padrão evidente. Já o espaço de parâmetros S² x K x $\alpha$ conseguiu alocar nos mesmos clusters sinais com comportamentos parecidos. Note que os ruídos branco e rosa pertencem ao grupo 0 e os mapas logístico e de henon pertencem ao grupo 1. Já os ruídos vermelho e estocástico necessitaram de dois outros cluster para alocar seus sinais.
